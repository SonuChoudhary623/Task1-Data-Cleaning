```python
df.columns=[col.lower().replace(' ','_') for col in df.columns]
```

```python
print(df.columns.tolist())
```

```python
print(df.isnull().sum())
```

```python
df['rating_count']=df['rating_count'].fillna('0')
```

```python
print(df.isnull().sum())
```

```python
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df['actual_price'] = (
    df['actual_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(int)
)
```

```python
df['rating_count'] = (
    df['rating_count']
    .str.replace(',', '', regex=False)
    .astype(int)
)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
print(df.dtypes)
```

```python
print('Duplicates:', df.duplicated().sum())
```

```python
df['main_category'] = df['category'].apply(
    lambda x: x.split('|')[0]
)
```
